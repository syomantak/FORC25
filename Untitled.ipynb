```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math
import matplotlib
```

```python
def s2n(z):
    numbers = z.split(' ')
    ns = []
    for number in numbers:
        if '×' in number:
            a = number.split('×10')
            ns.append(-np.log(float(a[0]+'E-'+a[1][1:])))
        else:
            ns.append(-np.log(float(number)))
        
    return ns

s2n('0.118 0.174 0.273 1.000 0.459 0.420')
```

```python
ucp = s2n('0.118 0.174 0.273 1.000 0.459 0.420')
uce = s2n('0.011 0.030 0.074 0.396 0.180 0.087')
fcp = s2n('0.105 0.083 0.202 1.000 0.924 0.253')
fce = s2n('0.012 0.007 0.041 0.834 0.779 0.027')
tcp = s2n('0.020 0.003 0.090 0.233 1.000 0.125')
tce = s2n('3×10−4 7×10−6 0.008 0.022 0.779 0.008')


Data_p = np.array([ucp,fcp,tcp])
Data_e = np.array([uce,fce,tce])
```

```python
width = 0.2
x = np.arange(3)*2 

patterns = [ "/" , "\\" , "|" , "-" , "+" , "x", "o", "O", ".", "*" ]

# plot data in grouped manner of bar type 
plt.bar(x-0.2, Data_p[:,0], width, color='#16001E', edgecolor='black' ) 
plt.bar(x, Data_p[:,1], width, color='#AA78A6', hatch = patterns[1], edgecolor='black') 
plt.bar(x+0.2, Data_p[:,2], width, color='#BD1E1E', hatch = patterns[8], edgecolor='black') 
plt.bar(x+0.4, Data_p[:,3], width, color='#F5E2C8', hatch = patterns[0], edgecolor='black') 
plt.bar(x+0.6, Data_p[:,4], width, color='#454ADE', edgecolor='black') 
plt.bar(x+0.8, Data_p[:,5], width, color='#35FF69', hatch = patterns[6], edgecolor='black') 


plt.xticks(x, [r'UC Dataset', r'$n=10^4,k=5$', r'$n=10^4,k=20$']) 
plt.xlabel("Experiments") 
plt.ylabel(" Negative log 95-th Quantile Error") 
plt.legend([r"HPF-C$\mathbb{P}$", "HPF-A", "Prop", 'UNI', 'SM', 'LDP']) 
plt.savefig("CorrPAC.pdf", 
            dpi=1000, 
            bbox_inches='tight', 
            )
plt.show() 
```
